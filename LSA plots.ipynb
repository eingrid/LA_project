```python
import json
import matplotlib.pyplot as plt
```

```python
topic_numbers = list(range(2,15)) + [20,30,40,50,100,150,200]
```

```python
with open('coherences_12grams.json','r') as json_file:
    coherences_en = json.load(json_file)['coherences']
```

```python
with open('coherences_12grams_pos.json','r') as json_file:
    coherences_en_pos = json.load(json_file)['coherences']
```

```python
with open('coherences_12grams_no_lemmatization.json','r') as json_file:
    coherences_en_no_lemmatization = json.load(json_file)['coherences']
```

```python
with open('coherences_12grams_uk.json','r') as json_file:
    coherences_uk = json.load(json_file)['coherences']
```

```python
with open('coherences_12grams_uk_new.json','r') as json_file:
    coherences_uk_new = json.load(json_file)['coherences']
```

```python
with open('coherences_12grams_uk_new_no_lemmatization.json','r') as json_file:
    coherences_uk_new_no_lemmatization = json.load(json_file)['coherences']
```

```python
with open('coherences_12grams_uk_new_pos.json','r') as json_file:
    coherences_uk_new_pos = json.load(json_file)['coherences']
```

```python
with open('coherences_12grams_uk_new_pos_top5.json','r') as json_file:
    coherences_uk_new_pos_top5 = json.load(json_file)['coherences']
```

```python
with open('coherences_12grams_uk_new_pos_nv.json','r') as json_file:
    coherences_uk_new_pos_nv = json.load(json_file)['coherences']
```

```python
with open('coherences_12grams_uk_new_pos_no_lemmatization.json','r') as json_file:
    coherences_uk_new_pos_no_lemmatization = json.load(json_file)['coherences']
```

# Plot 2. English and Ukrainian

```python
plt.figure(figsize=(9, 5))
plt.plot(topic_numbers[:-2], coherences_en[:-2], marker=".", color="blue", label='English')
plt.plot(topic_numbers[:-2], coherences_uk[:-2], marker="x", color="orange", label='Ukrainian')
plt.title(f'Coherence for different n_topics | Custom LSA | (1,2)-grams model | en & uk', fontsize=13)
plt.xlabel('Topics number', fontsize=13)
plt.ylabel('Coherence (c_v) score', fontsize=13)
plt.grid()
plt.legend(fontsize=13)
plt.savefig('language_comparison_plot.png')
plt.show()
```

# Plot 3. The Effect of Parts of Speech on Topic Modeling

```python
plt.figure(figsize=(9, 5))
plt.plot(topic_numbers[:-2], coherences_uk_new[:-2], marker=".", color="gray",
         label='All parts of speech (PoS)')
plt.plot(topic_numbers[:-2], coherences_uk_new_pos[:-2], marker="x", color="green",
         label='Selected PoS (NOUN, PNOUN, VERB, ADJ, ADV)')
# plt.plot(topic_numbers[:-2], coherences_uk_new_pos_top5[:-2], marker="x", color="lightgreen",
#          label='Selected PoS (NOUN, PNOUN, VERB, ADJ, ADV)')
plt.plot(topic_numbers[:-2], coherences_uk_new_pos_nv[:-2], marker="+", color="fuchsia",
         label='Selected PoS (NOUN, PNOUN, VERB)')
plt.title(f'Coherence for different n_topics | Custom LSA | (1,2)-grams model | uk w & w/o PoS', fontsize=13)
plt.xlabel('Topics number', fontsize=13)
plt.ylabel('Coherence (c_v) score', fontsize=13)
plt.grid()
plt.legend(fontsize=13)
plt.savefig('pos_ukrainian_plot_1.png')
plt.show()
```

# Plot 4. Influence of Various Preprocessing Techniques

```python
plt.figure(figsize=(9, 5))
plt.plot(topic_numbers[:-2], coherences_en[:-2], marker=".", color="blue", label='en')
plt.plot(topic_numbers[:-2], coherences_en_no_lemmatization[:-2], marker="x", color="lightblue",
         label='en, no lemmatization')
plt.plot(topic_numbers[:-2], coherences_uk_new[:-2], marker=".", color="grey",
         label='uk, all PoS')
plt.plot(topic_numbers[:-2], coherences_uk_new_no_lemmatization[:-2], marker="x", color="darkgrey",
         label='uk, all PoS, no lemmatization')
plt.plot(topic_numbers[:-2], coherences_uk_new_pos[:-2], marker=".", color="green",
         label='uk, 5 selected PoS')
plt.plot(topic_numbers[:-2], coherences_uk_new_pos_no_lemmatization[:-2], marker="x", color="lightgreen",
         label='uk, 5 selected PoS, no lemmatization')
plt.title(f'Coherence for different n_topics | Custom LSA | (1,2)-grams model | en & uk', fontsize=13)
plt.xlabel('Topics number', fontsize=13)
plt.ylabel('Coherence (c_v) score', fontsize=13)
plt.grid()
plt.legend(fontsize=13)
plt.savefig('preprocessing_ukrainian_plot_1.png')
plt.show()
```
